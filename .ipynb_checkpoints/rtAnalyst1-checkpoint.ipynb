## #2 HOUSEHOLDS MUST HAVE A VALID PHONE NUMBERS

```python
PhoneNumber = data['SĐT']
```

```python
data=data[PhoneNumber>0]
data
```

##### In your dataset, All Phone numbers are mobile phone which i just need to include '0' at the beginning of these numbers. There's no land-line so i dont need to include zone code (like 024 or 028) here

```python
AfterConvert=[]
for i in data['SĐT']:
    i=i.astype(str)
    k=i.replace(i,('0'+i))
    #j=pd.to_numeric(k, errors='coerce')
    AfterConvert.append(k)
```

```python
data.insert(5,'SĐT Mới',value=AfterConvert)
```

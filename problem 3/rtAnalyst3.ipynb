# #1 Number of interviews each numerator has completed!

```python
def to_frequency_table(data):
    frequencytable = {}
    for key in data:
        if key in frequencytable:
            frequencytable[key] += 1
        else:
            frequencytable[key] = 1
    return frequencytable
```

```python
to_frequency_table(data.entry)
```
